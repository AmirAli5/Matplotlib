# Pie Charts
A pie chart is best used when we trying to work out the composition of something. If we have categorical data then using a pie chart would work really well as each slice(like response in our example) can represent a different category. 

#### Import Library

```python
from matplotlib import pyplot as plt
```

#### Create Dataset

```python
response = [59219, 55466, 47544, 36443, 35917]
languages = [ 'JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

### Create Pie Chart

```python
plt.pie(response, labels = languages, shadow = True, wedgeprops = {'edgecolor': 'black'}, autopct = '%1.1f%%', startangle= 90)
plt.title("Response on Languages")
plt.tight_layout()
plt.show()
```

```python
# Apply fivethirtyeight style
plt.style.use('fivethirtyeight')
plt.pie(response, labels = languages, shadow = True, wedgeprops = {'edgecolor': 'black'}, autopct = '%1.1f%%', startangle= 90)
plt.title("Response on Languages") 
plt.tight_layout()
plt.show()
```

```python
# Explode for python
explode = [0, 0, 0, 0.1, 0]
plt.pie(response, labels = languages, shadow = True, explode = explode, wedgeprops = {'edgecolor': 'black'}, autopct = '%1.1f%%', startangle= 90)
plt.title("Response on Languages")
plt.savefig("Graph/3.Pie Charts.png")
plt.tight_layout()
plt.show()
```
