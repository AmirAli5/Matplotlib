# Plotting Time Series
The time-series plot is a univariate plot: it shows only one variable. It is a 2-dimensional plot in which one axis, the time-axis, shows graduations at an appropriate scale (seconds, minutes, weeks, quarters, years), while the other axis shows the numeric values.

### Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Import Dataset

```python
dataset = pd.read_csv("C:\\Users\\amira\\Data Science\\0. Some  Basic Work\\4. Matplotlib\\Data\\bitcoindata.csv")
```

```python
Date = dataset["Date"]
Price_Close = dataset["Close"]
```

### Plotting Time Series Data

```python
plt.plot_date(Date, Price_Close, linestyle = 'solid')

plt.gcf().autofmt_xdate()

plt.title("Bitcoin Prices")
plt.ylabel("Closing price")
plt.xlabel("Date")

plt.tight_layout()

plt.savefig("Graph/8.Time Series.png")

plt.style.use("fivethirtyeight")

plt.show
```
