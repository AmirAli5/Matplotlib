# Stack Plots
The idea of stack plots is to show “parts to a whole” over time; basically, it's like a pie-chart, only over time. Stack plots are mainly used to see various trends in variables over a specific period of time.

### Import Library

```python
from matplotlib import pyplot as plt
```

### Create a Dataset

```python
minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

player1 = [1, 1, 2, 2, 3, 4, 4, 4, 5]
player2 = [1, 2, 2, 2, 2, 3, 4, 4, 4]
player3 = [1, 1, 2, 2 ,2, 2 ,2 ,3, 3]
```

### Create a Stack Plots

#### Customizing the Stack Plot

```python
plt.stackplot(minutes, player1, player2, player3, labels = ['Player1', 'Player2', 'Player3'],  colors = ['#6d904f', '#fc4f30', '#008fd5'])
plt.legend(loc = 'upper left')
plt.title("Players Distribution")
plt.xlabel('minutes')
plt.ylabel('Total Score')
plt.tight_layout()
plt.show()
```

#### Apply the fivethirtyeight Style

```python
plt.style.use('fivethirtyeight')
plt.stackplot(minutes, player1, player2, player3, labels = ['Player1', 'Player2', 'Player3'],  colors = ['#6d904f', '#fc4f30', '#008fd5'])
plt.legend(loc = 'upper left')
plt.title("Players Distribution")
plt.xlabel('minutes')
plt.ylabel('Total Score')
plt.savefig("Graph/4.Stack Plots.png") 
plt.tight_layout()
plt.show()
```
